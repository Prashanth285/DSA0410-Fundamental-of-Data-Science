```python
import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
temperature = [4, 6, 10, 15, 20, 25, 28, 27, 22, 16, 9, 5]
rainfall = [78, 60, 72, 55, 48, 35, 30, 40, 58, 70, 85, 90]

fig, ax1 = plt.subplots(figsize=(8, 5))

ax1.set_xlabel('Month')
ax1.set_ylabel('Temperature (°C)', color='red')
ax1.plot(months, temperature, marker='o', color='red', label='Temperature')
ax1.tick_params(axis='y', labelcolor='red')

ax2 = ax1.twinx()
ax2.set_ylabel('Rainfall (mm)', color='blue')
ax2.scatter(months, rainfall, color='blue', label='Rainfall')
ax2.tick_params(axis='y', labelcolor='blue')

plt.title('Monthly Temperature and Rainfall')
plt.grid(True)
plt.show()
```
